```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
simulation_time = 5 # Enter the desired simulation time
time = np.arange(0, simulation_time, 0.0001) 
sweep_time = 5 # Enter the duration of the sweep
```

## Sine Wave Generation

#### Amplitude of Sine Wave

```python
amplitude_sin = 100 # Enter the desired amplitude
```

#### Frequency of Sine Wave

```python
f_initial_sin = 1 # Enter the desired initial frequency
f_final_sin = 20 # Enter the desired final frequency
c_sin = (f_final_sin-f_initial_sin)/sweep_time
```

```python
sin = amplitude_sin*np.sin(2*np.pi*((c_sin/2)*time**2+f_initial_sin*time))
```

```python
plt.plot(time,sin)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Sine Wave');
```

#### Linear Chirp

```python
freq_sin = c_sin*time + f_initial_sin
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.title('Linear Chirp of Sine Frequency');
plt.plot(time,freq_sin);
```

## Triangle Wave Generation

#### Amplitude of Triangle Wave

```python
amplitude_tri = 100 # Enter the desired amplitude
```

#### Frequency of Triangle Wave

```python
f_initial_tri = 1  # Enter the desired initial frequency
f_final_tri = 20  # Enter the desired final frequency
c_tri = (f_final_tri-f_initial_tri)/sweep_time
```

```python
tri = amplitude_tri*signal.sawtooth(2*np.pi*((c_tri/2)*time**2+f_initial_tri*time))
```

```python
plt.plot(time,tri)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Triangle Wave');
```

#### Linear Chirp

```python
freq_tri = c_tri*time + f_initial_tri
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.title('Linear Chirp of Triangle Frequency');
plt.plot(time,freq_tri);
```

## Square Wave Generation

#### Amplitude of Square Wave

```python
amplitude_squ = 100 # Enter the desired amplitude
```

#### Frequency of Square Wave

```python
f_initial_squ = 1  # Enter the desired initial frequency
f_final_squ = 20 # Enter the desired final frequency
c_squ = (f_final_squ-f_initial_squ)/sweep_time
```

```python
squ = amplitude_squ*signal.square(2*np.pi*((c_squ/2)*time**2+f_initial_squ*time))
```

```python
plt.plot(time,squ)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Square Wave');
```

#### Linear Chirp

```python
freq_squ = c_squ*time + f_initial_squ
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.title('Linear Chirp of Square Frequency');
plt.plot(time,freq_squ);
```
